## This notebook applies the Absorption Feature Finder algorithm to the NSO spectrum

```python
import numpy as np
import matplotlib.pyplot as plt
import glob
import pandas as pd
```

```python
%matplotlib notebook
```

```python
template = pd.read_csv("NSOspec.csv")
keep = np.where((template.Flux.values > 0) &
                (template.Wavelength.values >= 4500) &
                (template.Wavelength.values <= 6800))[0]
```

```python
plt.plot(template.Wavelength.values[keep], template.Flux.values[keep], c='r')
plt.show()
```

```python
from spectra_functions import findabsorptionfeatures
```

```python
#practice = np.where((template.Wavelength.values[keep] >= 4500) & (template.Wavelength.values[keep] <= 5000))[0]
wvbounds, minwvs, minflxs, maxflxs = findabsorptionfeatures(template.Wavelength.values[keep], 
                                                            template.Flux.values[keep],
                                                            pix_range=25, minlinedepth=0.015, alpha=0.05, gamma=0.01)
```

```python
len(wvbounds)
```

```python
plt.plot(template.Wavelength.values[keep], template.Flux.values[keep])
for i in range(len(wvbounds)):
    if i%2 == 0:
        plt.plot(wvbounds[i], (1.05, 1.05), c='red')
    else:
        plt.plot(wvbounds[i], (1.03, 1.03), c='red')
plt.show()
```

```python
wvlbnds = []
wvubnds = []
minflxs2 = []
minwvs2 = []
maxflxs2 = []
for i in range(len(wvbounds)):
    rng = np.where((template.Wavelength.values >= wvbounds[i][0]) & (template.Wavelength.values <= wvbounds[i][1]))[0]
    nans = np.where(np.isnan(template.Flux.values[rng]))[0]
    if len(nans) == 0:
        wvlbnds.append(wvbounds[i][0])
        wvubnds.append(wvbounds[i][1])
        minflxs2.append(minflxs[i])
        minwvs2.append(minwvs[i])
        maxflxs2.append(maxflxs[i])
Features = pd.DataFrame({"Wv_lbounds": wvlbnds, "Wv_ubounds": wvubnds, "Minflux": minflxs2, "MinWvl": minwvs2, "Maxflux": maxflxs2})
Features.to_csv("Features.csv")
```
